```python
# Create User
def create_user(conn, first_name, last_name, age, gender, phone_no, city, state, email, password):
    cursor = conn.cursor()
    query = """
    INSERT INTO User (First_name, Last_name, Age, Gender, Phone_no, City, State, Email, Password)
    VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s)
    """
    cursor.execute(query, (first_name, last_name, age, gender, phone_no, city, state, email, password))
    conn.commit()
    print("User created successfully")

# Read Users
def read_users(conn):
    cursor = conn.cursor()
    query = "SELECT * FROM User"
    cursor.execute(query)
    result = cursor.fetchall()
    
    for user in result:
        print(user)

# Update User
def update_user(conn, user_id, first_name, last_name, age, phone_no, email):
    cursor = conn.cursor()
    query = """
    UPDATE User
    SET First_name = %s, Last_name = %s, Age = %s, Phone_no = %s, Email = %s
    WHERE User_ID = %s
    """
    cursor.execute(query, (first_name, last_name, age, phone_no, email, user_id))
    conn.commit()
    print("User updated successfully")

# Delete User
def delete_user(conn, user_id):
    cursor = conn.cursor()
    query = "DELETE FROM User WHERE User_ID = %s"
    cursor.execute(query, (user_id,))
    conn.commit()
    print("User deleted successfully")
```
